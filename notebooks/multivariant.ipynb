# Linear discriminant analysis 

```r
library("MASS")                                                # load the MASS package
wine.lda <- lda(V1 ~ V2 + V3 +  V4 + V5 + V6 + V7 + 
        V8 + V9 + V10 + V11 + V12 + V13 +V14,  data = wine)
wine.lda
```

```r
wine.lda.values = predict(wine.lda, wine[2:14])
names(wine.lda.values)
```

```r
wine.lda.values$class
```

```r
wine.lda.values$posterior
```
